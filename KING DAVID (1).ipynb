# LINE CHART: INCOME QUINTILES BY YEAR

```python
plt.figure(figsize=(10, 6))
plt.plot(wealth_inequality['Year'], wealth_inequality['quintile1'], label='quintile1')
plt.plot(wealth_inequality['Year'], wealth_inequality['quintile2'], label='quintile2')
plt.plot(wealth_inequality['Year'], wealth_inequality['quintile3'], label='quintile3')
plt.plot(wealth_inequality['Year'], wealth_inequality['quintile4'], label='quintile4')
plt.xlabel('Year')
plt.ylabel('Income')
plt.title('Income Inequality In the US Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
